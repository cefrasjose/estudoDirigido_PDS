# Simulacões Computacionais

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Definição do vetor de tempo discreto (n)

    n = np.arange(-10, 11)

### 2. Geração de Sequências Elementares

### Impulso Unitário

    impulso = np.where(n == 0, 1, 0)

### Degrau Unitário

    degrau = np.where(n >= 0, 1, 0)

### Exponencial Complexa (Amortecida) real para visualização

    exponencial = (0.8 ** n) * degrau

### 3. Plotagem dos Gráficos

    plt.figure(figsize=(12, 4))

```python
n = np.arange(-10, 11)

impulso = np.where(n == 0, 1, 0)

degrau = np.where(n >= 0, 1, 0)

exponencial = (0.8 ** n) * degrau

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.stem(n, impulso)
plt.title('Impulso Unitário $\delta[n]$')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.stem(n, degrau)
plt.title('Degrau Unitário $u[n]$')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.stem(n, exponencial)
plt.title('Exponencial Amortecida')
plt.grid(True)

plt.tight_layout()
plt.savefig('../resultados/sinais_elementares.png')
plt.show()
```
