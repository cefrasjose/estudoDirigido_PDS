# Simulacões Computacionais

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Geração de Sinais e Exponencial Complexa
    
    n = np.arange(-10, 11)

### Impulso e Degrau

    impulso = np.where(n == 0, 1, 0)
    degrau = np.where(n >= 0, 1, 0)

### Exponencial Complexa: x[n] = e^(j * w * n)

    omega = np.pi / 4
    exp_complexa = np.exp(1j * omega * n)

    
# 2. Cálculo de Energia e Potência
    
### Sinal de Energia (Duração finita / Amortecido)

    n_energia = np.arange(0, 20)
    sinal_energia = (0.7 ** n_energia) * np.where(n_energia >= 0, 1, 0)
    energia_total = np.sum(np.abs(sinal_energia)**2)

### Sinal de Potência (Duração infinita / Periódico)

    n_potencia = np.arange(0, 100)
    sinal_potencia = 3 * np.cos(np.pi / 4 * n_potencia)
    potencia_media = np.mean(np.abs(sinal_potencia)**2)

    print(f"Cálculo de Energia (Pulso Amortecido): E = {energia_total:.4f} Joules (unidade teórica)")
    print(f"Cálculo de Potência (Sinal Periódico): P = {potencia_media:.4f} Watts (unidade teórica)")

    
# 3. Plotagem dos Gráficos
    
    plt.figure(figsize=(14, 8))

### Plot Exponencial Complexa (Parte Real e Imaginária)

    plt.subplot(2, 2, 1)
    plt.stem(n, np.real(exp_complexa), linefmt='b-', markerfmt='bo', basefmt='k-')
    plt.title('Exponencial Complexa (Parte Real)')
    plt.grid(True)

    plt.subplot(2, 2, 2)
    plt.stem(n, np.imag(exp_complexa), linefmt='r-', markerfmt='ro', basefmt='k-')

    plt.grid(True)

    # Plot Sinal de Energia
    plt.subplot(2, 2, 3)

    plt.grid(True)

# Plot Sinal de Potência (Apenas um trecho para visualização)

    plt.subplot(2, 2, 4)
    plt.stem(n_potencia[:20], sinal_potencia[:20], linefmt='m-', markerfmt='mo', basefmt='k-')

```python
n = np.arange(-10, 11)

impulso = np.where(n == 0, 1, 0)
degrau = np.where(n >= 0, 1, 0)

omega = np.pi / 4
exp_complexa = np.exp(1j * omega * n)

n_energia = np.arange(0, 20)
sinal_energia = (0.7 ** n_energia) * np.where(n_energia >= 0, 1, 0)
energia_total = np.sum(np.abs(sinal_energia)**2)

n_potencia = np.arange(0, 100)
sinal_potencia = 3 * np.cos(np.pi / 4 * n_potencia)
potencia_media = np.mean(np.abs(sinal_potencia)**2)

print(f"Cálculo de Energia (Pulso Amortecido): E = {energia_total:.4f} Joules (unidade teórica)")
print(f"Cálculo de Potência (Sinal Periódico): P = {potencia_media:.4f} Watts (unidade teórica)")

plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.stem(n, np.real(exp_complexa), linefmt='b-', markerfmt='bo', basefmt='k-')
plt.title('Exponencial Complexa (Parte Real)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.stem(n, np.imag(exp_complexa), linefmt='r-', markerfmt='ro', basefmt='k-')
plt.title('Exponencial Complexa (Parte Imaginária)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.stem(n_energia, sinal_energia, linefmt='g-', markerfmt='go', basefmt='k-')
plt.title(f'Sinal de Energia (E = {energia_total:.2f})')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.stem(n_potencia[:20], sinal_potencia[:20], linefmt='m-', markerfmt='mo', basefmt='k-')
plt.title(f'Sinal de Potência (P = {potencia_media:.2f})')
plt.grid(True)

plt.tight_layout()
plt.savefig('../resultados/analise_sinais.png')
plt.show()
```
